```python
box_office =[]
titles = []
pop_score = []
crit_score =[]
rating = []
genre = []
director = []
producer = []
release = []
box_office = []
runtime = []
for i in range(0,10):
    new_Url = soup.findAll('div',attrs={'class' : 'article_movie_title'})[i].find('a').get('href')
    new_response = requests.get(new_Url)
    new_soup = BS(new_response.text)
    Titles = (soup.findAll('div',attrs={'class' : 'article_movie_title'})[i].find('a'))
    titles.append(Titles)
    new_Url = soup.findAll('div',attrs={'class' : 'article_movie_title'})[i].find('a').get('href')
    new_response = requests.get(new_Url)
    new_soup = BS(new_response.text)
    Pop_Score = new_soup.findAll('div',attrs={'id' : 'topSection'})[0].findAll('div',attrs={'class' : 'thumbnail-scoreboard-wrap'})[0].findAll('score-board')[0]['audiencescore']
    pop_score.append(Pop_Score)
    Crit_Score = new_soup.findAll('div',attrs={'id' : 'topSection'})[0].findAll('div',attrs={'class' : 'thumbnail-scoreboard-wrap'})[0].findAll('score-board')[0]['tomatometerscore']
    crit_score.append(Crit_Score)
    for t in range (0,10):
            if new_soup.findAll('div',attrs={'class' : 'panel-body content_body'})[0].findAll('div',attrs={'class':'meta-label'})[t].text =='Release Date (Theaters):':
                Release = new_soup.findAll('div',attrs={'class' : 'panel-body content_body'})[0].findAll('div',attrs={'class':'meta-value'})[t].findAll('time')[0].text
                release.append(Release)
            elif new_soup.findAll('div',attrs={'class' : 'panel-body content_body'})[0].findAll('div',attrs={'class':'meta-label'})[t].text =='Rating:':
                Rating = (new_soup.findAll('div',attrs={'class' : 'panel-body content_body'})[0].findAll('div',attrs={'class':'meta-value'})[t].text[0:5])
                rating.append(Rating)
            elif new_soup.findAll('div',attrs={'class' : 'panel-body content_body'})[0].findAll('div',attrs={'class':'meta-label'})[t].text =='Genre:':
                Genre = (new_soup.findAll('div',attrs={'class' : 'panel-body content_body'})[0].findAll('div',attrs={'class':'meta-value'})[t].text[50:56])
                genre.append(Genre)
            elif new_soup.findAll('div',attrs={'class' : 'panel-body content_body'})[0].findAll('div',attrs={'class':'meta-label'})[t].text =='Director:':
                Director = (new_soup.findAll('div',attrs={'class' : 'panel-body content_body'})[0].findAll('div',attrs={'class':'meta-value'})[t].findAll('a')[0].text)
                director.append(Director)
            elif new_soup.findAll('div',attrs={'class' : 'panel-body content_body'})[0].findAll('div',attrs={'class':'meta-label'})[t].text =='Producer:':
                Producer = (new_soup.findAll('div',attrs={'class' : 'panel-body content_body'})[0].findAll('div',attrs={'class':'meta-value'})[t].findAll('a')[0].text)
                producer.append(Producer)
            #elif new_soup.findAll('div',attrs={'class' : 'panel-body content_body'})[0].findAll('div',attrs={'class':'meta-label'})[t].text =='Box Office (Gross USA):':
                #Box_Office = (new_soup.findAll('div',attrs={'class' : 'panel-body content_body'})[0].findAll('div',attrs={'class':'meta-value'})[t].text)
                #box_office.append(Box_Office)
            elif new_soup.findAll('div',attrs={'class' : 'panel-body content_body'})[0].findAll('div',attrs={'class':'meta-label'})[t].text =='Runtime:':
                Runtime = (new_soup.findAll('div',attrs={'class' : 'panel-body content_body'})[0].findAll('div',attrs={'class':'meta-value'})[t].text[30:36])
                runtime.append(Runtime)
    if new_soup(text = "Box Office (Gross USA):"):
        value = (
                new_soup(text = "Box Office (Gross USA):")[0] # grab the text item
                .parent # find the parent (tag) of the text item
                .find_next('div')   # find the next div tag
                .text   # get the text
                )
    else:
        value = None
    box_office.append(value)
box_office
```

```python
rtom_rating = pd.DataFrame(rating)
rtom_rating = rtom_rating.rename(columns={ rtom_rating.columns[0]: "Rating" })
rtom_genre = pd.DataFrame(genre)
rtom_genre = rtom_genre.rename(columns={ rtom_genre.columns[0]: "Genre" })
rtom_director = pd.DataFrame(director)
rtom_director = rtom_director.rename(columns={ rtom_director.columns[0]: "Director" })
rtom_producer = pd.DataFrame(producer)
rtom_producer = rtom_producer.rename(columns={ rtom_producer.columns[0]: "Producer" })
rtom_release = pd.DataFrame(release)
rtom_release = rtom_release.rename(columns={ rtom_release.columns[0]: "Release Date" })
rtom_box_office = pd.DataFrame(box_office)
rtom_box_office = rtom_box_office.rename(columns={ rtom_box_office.columns[0]: "Box Office" })
rtom_runtime = pd.DataFrame(runtime)
rtom_runtime = rtom_runtime.rename(columns={ rtom_runtime.columns[0]: "Runtime" })
```
